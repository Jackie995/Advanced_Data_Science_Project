```python
import pandas as pd

df = pd.read_csv("visa issuance.csv")

df = df.set_index('Issuing Office')

print(df.head())
```

```python
df.dropna()
```

```python
# There's so many issuing offices, let's merge them and sort them by country!
df_country = df.copy()
df_country['Country'] = df_country.index.str.split(',').str[0]
df_aggregated = df_country.groupby('Country').sum()
df_clean = df_aggregated.loc[~(df_aggregated == 0).all(axis=1)]
df_clean.shape
```

```python
df_clean
```

```python
# Export Dataframe into CSV
df_clean.to_csv('visa_issuance.csv')
```
